```python
import numpy as np 
import pandas as pd
Netflix = pd.read_csv('./NTFLX.csv')
Netflix.info()
```

```python
Netflix.head()
```

```python
Netflix['Date'] = pd.to_datetime(Netflix['Date'])
print(Netflix)
print (Netflix.dtypes)
```

```python
Netflix.set_index('Date', inplace=True)
Netflix.head()
```

```python
Netflix. drop('Company', axis=1, inplace=True)
Netflix
```

```python
import matplotlib.pyplot as plt
Netflix.plot(subplots=True, figsize=(10,12))
plt.title('Netflix stock attributes')
plt.savefig('stocks.png')
plt.show()
```

```python
Netflix['2020':'2021'].plot(subplots=True, figsize=(10,12))
plt.title('Netflix stock attributes')
plt.savefig('stocks.png')
plt.show()
```
